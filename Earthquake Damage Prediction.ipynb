## Standarisasi

```python
from sklearn.preprocessing import MinMaxScaler
# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

# Fit dan transform data train
train_encoded[numerical_columns] = scaler.fit_transform(train_encoded[numerical_columns])

# Transform data test dengan parameter yang sama dari data train
test_encoded[numerical_columns] = scaler.transform(test[numerical_columns])
```

# Splitting Data

```python
X = train_encoded.drop(['damage_grade', 'ID'],axis = 1)
y = train_encoded['damage_grade']
```

```python
# Split data menjadi train dan val
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state = 26)
```

# SMOTE

```python
from imblearn.over_sampling import SMOTE
# Inisialisasi objek SMOTE
smote = SMOTE(random_state=21725)

# Melakukan oversampling pada data
X_resampled, y_resampled = smote.fit_resample(X_train,y_train)
```

# Pemodelan

```python
# Import library atas algoritma / metode klasifikasi yang digunakan
from sklearn.ensemble import HistGradientBoostingClassifier, GradientBoostingClassifier, RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from catboost import CatBoostClassifier

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report
```

## Cat Boost

```python
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report
cat_clf = CatBoostClassifier()
cat_clf.fit(X_resampled, y_resampled)
cat_pred = cat_clf.predict(X_val)
metrics = classification_report(y_val, cat_pred, digits=4)
print("\n",metrics)
```

```python
cek = test_encoded.copy()
cek.set_index(['ID'], inplace = True)
cek['damage_grade'] = cat_clf.predict(cek)
cek['ID'] = cek.index
cat = cek[['ID', 'damage_grade']]
cat
```
